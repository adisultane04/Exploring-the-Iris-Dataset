```python
df = sns.load_dataset("iris")
```

## **Check the balance of the target variable (species count)**

```python
df['Species'].value_counts()
```

## **Create histograms and box plots for each numerical feature**

```python
sns.histplot(df['SepalLengthCm'], kde=True)
```

```python
sns.histplot(df['SepalWidthCm'], kde=True)
```

```python
sns.histplot(df['PetalLengthCm'], kde=True)
```

```python
sns.histplot(df['PetalWidthCm'], kde=True)
```

## **Create box plots for each numerical feature**

```python
# Boxplots (all features together)
sns.boxplot(data=df)
plt.title("Boxplot of Numerical Features")
plt.show()
```

```python
sns.pairplot(df,hue="Species", diag_kind="hist")
plt.show()
```

### **Use a pairplot to visualize relationships between all features at once**

```python
sns.pairplot(df, hue="Species")
plt.show()
```

### **Compute a correlation matrix and display it as a heatmap**

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```

```python
sns.countplot(data=df, x='Species', hue='Species')
plt.show()
```
